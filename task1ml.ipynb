```python
import pandas as pd
```

```python
data_train = pd.read_csv('train.csv')
data_test = pd.read_csv('test.csv')
```

```python
data_train.info()
```

```python
data_train
```

```python
data_test
```

```python
features = ['GrLivArea', 'BedroomAbvGr', 'FullBath', 'HalfBath']
target = 'SalePrice'
```

```python
data_train = data_train[features + [target]].dropna()
```

```python
x = data_train[features]
y = data_train[target]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_data_train, X_data_test, y_data_train, y_data_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
X_data_train
```

```python
X_data_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_data_train, y_data_train)
```

```python
y_pred = model.predict(X_data_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error
```

```python
mse = mean_squared_error(y_data_test, y_pred)
mae = mean_absolute_error(y_data_test, y_pred)
print(f'Mean absolute Error: {mae}')
print(f'Mean Squared Error: {mse}')
```

```python
from sklearn.metrics import r2_score

```

```python
r2 = r2_score(y_data_test, y_pred)

print(f'R^2 Score: {r2}')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(10, 6))
plt.scatter(y_data_test, y_pred, c='blue', label='Predicted vs Actual')
plt.plot([min(y_data_test), max(y_data_test)], [min(y_data_test), max(y_data_test)], color='gray', linestyle=':', linewidth=2, label='Perfect Prediction Line')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual Prices vs. Predicted Prices with Regression Line')
plt.legend()
plt.show()
```

```python
results_df = pd.DataFrame({'Actual Prices': y_data_test, 'Predicted Prices': y_pred})
results_df.to_csv('predictions.csv', index=False)

# Load the test data for new predictions
test_data = pd.read_csv('test.csv')

# Define the features for prediction
features = ['GrLivArea', 'BedroomAbvGr', 'FullBath', 'HalfBath']

# Prepare the test data
X_n = test_data[features]

```
